```python
import pandas as pd
import glob
```

```python
df1 = pd.read_csv('2023_data/202301.csv', encoding='euc-kr')
df2 = pd.read_csv('2023_data/202302.csv', encoding='euc-kr')
df3 = pd.read_csv('2023_data/202303.csv', encoding='euc-kr')
df4 = pd.read_csv('2023_data/202304.csv', encoding='euc-kr')
df5 = pd.read_csv('2023_data/202305.csv', encoding='euc-kr')
df6 = pd.read_csv('2023_data/202306.csv', encoding='euc-kr')
```

```python
mapping_data = pd.read_excel('서울생활이동데이터_행정동코드_20210907.xlsx')
```

```python
df1.head()
```

```python
mapping_data.head()
```

```python
mapping_data = mapping_data[['읍면동', 'full_name']]
```

```python
code_to_name = dict(zip(mapping_data['읍면동'], mapping_data['full_name']))
```

```python
df_list = [df1, df2, df3, df4, df5, df6]
```

```python
for i in range(len(df_list)):
    df_list[i]['행정동명'] = df_list[i]['도착 행정동 코드'].map(code_to_name)
    
    df_list[i] = df_list[i].drop(columns=['도착 행정동 코드'])

    print(f"df{i+1} 변환 결과:")
    print(df_list[i].head())
```

```python
for i, df in enumerate(df_list, start=1):

    if '도착 행정동 코드' in df.columns:
        df = df.drop(columns=['도착 행정동 코드'])
    
    # 데이터프레임을 업데이트
    df_list[i - 1] = df
    
    file_name = f"2023_data/df{i}.csv"  
    df.to_csv(file_name, index=False, encoding='euc-kr') 
    print(f"{file_name} 파일로 저장 완료")
```
