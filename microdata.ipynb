```python
import pandas as pd
```

```python
df_set = pd.read_excel('microdata_2023_집합.xlsx')
df_small = pd.read_excel('microdata_2023_소규모상가.xlsx')
df_medium = pd.read_excel('microdata_2023_중대형상가.xlsx')
```

```python
set_columns = [
    '상권명', '대지면적', '건축면적', '연면적', '전용면적(호)', '공용면적(호)', '임차인별면적합계',
    '제1월계약임대료_월세(총액)', '제2월계약임대료_월세(총액)', '제3월계약임대료_월세(총액)',
    '제1월시장임대료_㎡당월세임대료', '제2월시장임대료_㎡당월세임대료', '제3월시장임대료_㎡당월세임대료',
    '제1월공실', '제2월공실', '제3월공실', '임차인업종대분류', '임차인업종소분류'
]

small_medium_columns = [
    '상권명', '대지면적', '전용면적(호)', '공용면적(호)', '임차인별면적합계',
    '제1월계약임대료_월세(총액)', '제2월계약임대료_월세(총액)', '제3월계약임대료_월세(총액)',
    '제1월시장임대료_㎡당월세임대료', '제2월시장임대료_㎡당월세임대료', '제3월시장임대료_㎡당월세임대료',
    '제1월공실', '제2월공실', '제3월공실', '임차인업종대분류', '임차인업종소분류'
]

df_set = df_set[set_columns]
df_small = df_small[small_medium_columns]
df_medium = df_medium[small_medium_columns]

def convert_to_numeric(df, columns):
    for col in columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

numeric_columns = [
    '제1월계약임대료_월세(총액)', '제2월계약임대료_월세(총액)', '제3월계약임대료_월세(총액)',
    '제1월시장임대료_㎡당월세임대료', '제2월시장임대료_㎡당월세임대료', '제3월시장임대료_㎡당월세임대료',
    '제1월공실', '제2월공실', '제3월공실', '전용면적(호)', '공용면적(호)'
]

for df in [df_set, df_small, df_medium]:
    convert_to_numeric(df, numeric_columns)

# 1. 월세 상승률 계산 (3월 - 1월 / 1월 * 100)
for df in [df_set, df_small, df_medium]:
    df['월세상승률'] = ((df['제3월계약임대료_월세(총액)'] - df['제1월계약임대료_월세(총액)']) / df['제1월계약임대료_월세(총액)']) * 100
    
for df in [df_set, df_small, df_medium]:
    df['㎡당월세상승률'] = ((df['제3월시장임대료_㎡당월세임대료'] - df['제1월시장임대료_㎡당월세임대료']) / df['제1월시장임대료_㎡당월세임대료']) * 100

for df in [df_set, df_small, df_medium]:
    df['공실변화율'] = ((df['제3월공실'] - df['제1월공실']) / df['제1월공실']) * 100

for df in [df_set, df_small, df_medium]:
    df['상가면적합계'] = df['전용면적(호)'] + df['공용면적(호)']

df_set.head(), df_small.head(), df_medium.head()
```

```python
# 각 데이터프레임을 CSV 파일로 저장
df_set.to_csv('microdata_2023_set.csv', index=False, encoding='utf-8-sig')
df_small.to_csv('microdata_2023_small.csv', index=False, encoding='utf-8-sig')
df_medium.to_csv('microdata_2023_medium.csv', index=False, encoding='utf-8-sig')
```
