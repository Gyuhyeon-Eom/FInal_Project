```python
df_yes = pd.read_csv('최종 데이터(yes service).csv')
df_no = pd.read_csv('최종 데이터(no service).csv')
```

```python
# 서비스 대분류 생성 함수 수정 및 적용
def categorize_service(service_name):
    if '학원' in service_name or '독서실' in service_name or '컴퓨터학원' in service_name:
        return '교육업'
    elif ('슈퍼마켓' in service_name or '편의점' in service_name or '판매' in service_name or
          '의약품' in service_name or '서적' in service_name or '화장품' in service_name or 
          '가방' in service_name or '시계' in service_name or '핸드폰' in service_name or
          '중고차판매' in service_name or '가전제품' in service_name or '가구' in service_name or
          '청과상' in service_name or '육류' in service_name or '수산물' in service_name):
        return '소매업'
    elif ('음식점' in service_name or '커피' in service_name or '음료' in service_name or 
          '제과점' in service_name or '패스트푸드' in service_name or '분식' in service_name or
          '치킨' in service_name or '호프' in service_name or '간이주점' in service_name):
        return '음식업'
    elif ('의원' in service_name or '병원' in service_name or '의료' in service_name or
          '동물병원' in service_name or '한의원' in service_name):
        return '의료업'
    elif ('부동산' in service_name or '법무' in service_name or '여행사' in service_name or
          '통번역' in service_name or '청소' in service_name or '세무' in service_name or
          '변호사' in service_name or '회계사' in service_name or '변리사' in service_name):
        return '서비스업'
    elif ('PC방' in service_name or '당구장' in service_name or '골프' in service_name or 
          '스포츠' in service_name or '노래방' in service_name or '볼링장' in service_name or
          'DVD방' in service_name or '비디오' in service_name or '오락장' in service_name):
        return '레저/여가업'
    else:
        return '기타'

# 새로운 열 추가
df_yes['major'] = df_yes['service'].apply(categorize_service)
```

```python
gentrification_categories = ['음식업', '소매업', '서비스업']
df_gentrification = df_yes[df_yes['major'].isin(gentrification_categories)]
```

```python
df_gentrification
```

```python
df_gentrification = df_yes.copy()

# '음식업'과 '소매업'만을 사용하여 LQ 지수를 계산하고 필요한 파생 변수를 생성

# 음식업 데이터 필터링 및 외식업 점포 비율 LQ 지수 계산
food_service_df = df_gentrification[df_gentrification['major'] == '음식업']
food_service_rate = food_service_df.groupby(['dong', 'quarter'])['store_count'].sum().reset_index(name='음식업_점포_수')
food_service_total = food_service_df.groupby('quarter')['store_count'].sum().reset_index(name='전체_음식업_점포_수')
food_service_rate = pd.merge(food_service_rate, food_service_total, on='quarter', how='left')
food_service_rate['food_service_lq'] = (food_service_rate['음식업_점포_수'] / food_service_rate['전체_음식업_점포_수']) / (
    food_service_rate['음식업_점포_수'].sum() / food_service_rate['전체_음식업_점포_수'].sum())

# 소매업 데이터 필터링 및 소매업 점포 비율 LQ 지수 계산
retail_service_df = df_gentrification[df_gentrification['major'] == '소매업']
retail_service_rate = retail_service_df.groupby(['dong', 'quarter'])['store_count'].sum().reset_index(name='소매업_점포_수')
retail_service_total = retail_service_df.groupby('quarter')['store_count'].sum().reset_index(name='전체_소매업_점포_수')
retail_service_rate = pd.merge(retail_service_rate, retail_service_total, on='quarter', how='left')
retail_service_rate['retail_service_lq'] = (retail_service_rate['소매업_점포_수'] / retail_service_rate['전체_소매업_점포_수']) / (
    retail_service_rate['소매업_점포_수'].sum() / retail_service_rate['전체_소매업_점포_수'].sum())

# 프랜차이즈 비율 LQ 지수 계산 (모든 데이터 사용)
franchise_rate = df_gentrification.groupby(['dong', 'quarter'])['franchise_count'].sum().reset_index(name='프랜차이즈_수')
franchise_total = df_gentrification.groupby('quarter')['franchise_count'].sum().reset_index(name='전체_프랜차이즈_수')
franchise_rate = pd.merge(franchise_rate, franchise_total, on='quarter', how='left')
franchise_rate['franchise_lq'] = (franchise_rate['프랜차이즈_수'] / franchise_rate['전체_프랜차이즈_수']) / (
    franchise_rate['프랜차이즈_수'].sum() / franchise_rate['전체_프랜차이즈_수'].sum())

# 청년층 매출액 비율 LQ 지수 계산 (모든 데이터 사용)
df_gentrification['청년층_매출'] = (
    df_gentrification['age_30s_sales_amount'] +
    df_gentrification['age_20s_sales_amount'] +
    df_gentrification['age_30s_sales_amount']
)

total_sales_per_quarter = df_gentrification.groupby(['dong', 'quarter'])['청년층_매출'].sum().reset_index(name='청년층_매출_비율')
total_sales_overall = df_gentrification.groupby('quarter')['청년층_매출'].sum().reset_index(name='전체_청년층_매출')

# 데이터 병합
total_sales_per_quarter = pd.merge(total_sales_per_quarter, total_sales_overall, on='quarter', how='left')

# 청년층 매출 비율 LQ 지수 계산
total_sales_per_quarter['youth_sales_lq'] = (total_sales_per_quarter['청년층_매출_비율'] / total_sales_per_quarter['전체_청년층_매출']) / (
    total_sales_per_quarter['청년층_매출_비율'].sum() / total_sales_per_quarter['전체_청년층_매출'].sum()
)

# 개업률과 폐업률 변수 활용하여 개폐업 비율 계산 및 LQ 지수 생성

# 개폐업 비율 계산 (개업률 - 폐업률)
df_gentrification['개폐업_비율'] = df_gentrification['opening_rate'] - df_gentrification['closure_rate']

# 개폐업 비율 LQ 지수 계산
open_close_rate = df_gentrification.groupby(['dong', 'quarter'])['개폐업_비율'].sum().reset_index(name='개폐업_비율_합계')
open_close_total = df_gentrification.groupby('quarter')['개폐업_비율'].sum().reset_index(name='전체_개폐업_비율')

# 데이터 병합
open_close_rate = pd.merge(open_close_rate, open_close_total, on='quarter', how='left')

# 개폐업 비율 LQ 지수 계산
open_close_rate['open_close_lq'] = (open_close_rate['개폐업_비율_합계'] / open_close_rate['전체_개폐업_비율']) / (
    open_close_rate['개폐업_비율_합계'].sum() / open_close_rate['전체_개폐업_비율'].sum()
)

# 필요한 열을 모아서 최종 데이터프레임 생성 (음식업, 소매업, 프랜차이즈, 청년층 매출액, 개폐업 비율 LQ 포함)
final_df = food_service_rate[['dong', 'quarter', 'food_service_lq']].merge(
    retail_service_rate[['dong', 'quarter', 'retail_service_lq']],
    on=['dong', 'quarter'], how='outer'
).merge(
    franchise_rate[['dong', 'quarter', 'franchise_lq']],
    on=['dong', 'quarter'], how='outer'
).merge(
    total_sales_per_quarter[['dong', 'quarter', 'youth_sales_lq']],
    on=['dong', 'quarter'], how='outer'
).merge(
    open_close_rate[['dong', 'quarter', 'open_close_lq']],
    on=['dong', 'quarter'], how='outer'
)
```
