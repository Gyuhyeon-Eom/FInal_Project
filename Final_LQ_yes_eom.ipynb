```python
df_yes = pd.read_csv('table_yes_service.csv')
df_no = pd.read_csv('최종 데이터(no service).csv')
```

```python
# 서비스 대분류 생성 함수 수정 및 적용
def categorize_service(service_name):
    if '학원' in service_name or '독서실' in service_name or '컴퓨터학원' in service_name:
        return '교육업'
    elif ('슈퍼마켓' in service_name or '편의점' in service_name or '판매' in service_name or
          '의약품' in service_name or '서적' in service_name or '화장품' in service_name or 
          '가방' in service_name or '시계' in service_name or '핸드폰' in service_name or
          '중고차판매' in service_name or '가전제품' in service_name or '가구' in service_name or
          '청과상' in service_name or '육류' in service_name or '수산물' in service_name):
        return '소매업'
    elif ('음식점' in service_name or '커피' in service_name or '음료' in service_name or 
          '제과점' in service_name or '패스트푸드' in service_name or '분식' in service_name or
          '치킨' in service_name or '호프' in service_name or '간이주점' in service_name):
        return '음식업'
    elif ('의원' in service_name or '병원' in service_name or '의료' in service_name or
          '동물병원' in service_name or '한의원' in service_name):
        return '의료업'
    elif ('부동산' in service_name or '법무' in service_name or '여행사' in service_name or
          '통번역' in service_name or '청소' in service_name or '세무' in service_name or
          '변호사' in service_name or '회계사' in service_name or '변리사' in service_name):
        return '서비스업'
    elif ('PC방' in service_name or '당구장' in service_name or '골프' in service_name or 
          '스포츠' in service_name or '노래방' in service_name or '볼링장' in service_name or
          'DVD방' in service_name or '비디오' in service_name or '오락장' in service_name):
        return '레저/여가업'
    else:
        return '기타'

# 새로운 열 추가
df_yes['major'] = df_yes['service'].apply(categorize_service)
```

```python
gentrification_categories = ['음식업', '소매업']
df_gentrification = df_yes[df_yes['major'].isin(gentrification_categories)]
```

```python
df_gentrification = df_yes.copy()
```

```python
# '음식업'과 '소매업'만을 사용하여 LQ 지수를 계산하고 필요한 파생 변수를 생성
# 먼저 특정 분기 서울시 전체 점포 수와 행정동별 점포수를 계산

df_gentrification['seoul_total_store_cnt'] = df_gentrification.groupby(['quarter'])['store_count'].transform('sum')
df_gentrification['dong_total_store_cnt'] = df_gentrification.groupby(['quarter', 'dong'])['store_count'].transform('sum')
df_gentrification.head(50)
```

```python
# 음식업 데이터 필터링 및 외식업 점포 비율 LQ 지수 계산
food_service_df = df_gentrification[df_gentrification['major'] == '음식업'].copy()

food_service_df['food_store_cnt'] = food_service_df.groupby(['quarter', 'dong'])['store_count'].transform('sum')
food_service_df['food_store_total_cnt'] = food_service_df.groupby('quarter')['store_count'].transform('sum')
food_service_df['food_service_lq'] = (food_service_df['food_store_cnt'] / food_service_df['dong_total_store_cnt']) / (
food_service_df['food_store_total_cnt'] / food_service_df['seoul_total_store_cnt'])

food_service_df
```

```python
# 소매업 데이터 필터링 및 소매업 점포 비율 LQ 지수 계산
retail_df = df_gentrification[df_gentrification['major'] == '소매업'].copy()

retail_df['food_store_cnt'] = retail_df.groupby(['quarter', 'dong'])['store_count'].transform('sum')
retail_df['food_store_total_cnt'] = retail_df.groupby('quarter')['store_count'].transform('sum')
retail_df['retail_lq'] = (retail_df['food_store_cnt'] / retail_df['dong_total_store_cnt']) / (
retail_df['food_store_total_cnt'] / retail_df['seoul_total_store_cnt'])

retail_df
```

```python
# 프랜차이즈 비율 LQ 지수 계산 (모든 데이터 사용)

df_gentrification['seoul_total_frc_cnt'] = df_gentrification.groupby(['quarter'])['franchise_count'].transform('sum')
df_gentrification['dong_total_frc_cnt'] = df_gentrification.groupby(['quarter', 'dong'])['franchise_count'].transform('sum')
df_gentrification['franchise_lq'] = (df_gentrification['dong_total_frc_cnt'] / df_gentrification['dong_total_store_cnt']) / (
    df_gentrification['seoul_total_frc_cnt'] / df_gentrification['seoul_total_store_cnt'])

df_gentrification['franchise_lq']
```

```python
# 청년층 매출액 비율 LQ 지수 계산 (모든 데이터 사용)
df_gentrification['youth_sales_amount'] = (
    df_gentrification['age_10s_sales_amount'] +
    df_gentrification['age_20s_sales_amount'] +
    df_gentrification['age_30s_sales_amount']
)
df_gentrification['seoul_total_sales'] = df_gentrification.groupby(['quarter'])['current_month_sales_amount'].transform('sum')
df_gentrification['dong_total_sales'] = df_gentrification.groupby(['quarter', 'dong'])['current_month_sales_amount'].transform('sum')
df_gentrification['seoul_total_youth_sales'] = df_gentrification.groupby(['quarter'])['youth_sales_amount'].transform('sum')
df_gentrification['dong_total_youth_sales'] = df_gentrification.groupby(['quarter', 'dong'])['youth_sales_amount'].transform('sum')
df_gentrification['youth_sales_lq'] = (df_gentrification['dong_total_youth_sales'] / df_gentrification['dong_total_sales']) / (
    df_gentrification['seoul_total_youth_sales'] / df_gentrification['seoul_total_sales'])

df_gentrification['youth_sales_lq']
```

```python
grouped = df_gentrification.groupby(['dong', 'quarter'], as_index=False).agg({
    'open_store_count': 'sum',
    'closed_store_count': 'sum',
    'store_total_count': 'first'
})

grouped['total_net_growth_rate'] = (
    ((grouped['open_store_count'] / grouped['store_total_count']) -
     (grouped['closed_store_count'] / grouped['store_total_count'])) * 100
)

df_gentrification = df_gentrification.merge(
    grouped[['dong', 'quarter', 'total_net_growth_rate']],
    on=['dong', 'quarter'],
    how='left'
)
```

```python
df_gentrification = df_gentrification.drop(columns=['total_net_growth_rate_x'])
df_gentrification.rename(columns={'total_net_growth_rate_y': 'total_net_growth_rate'}, inplace=True)
```

```python
# 필요한 열 병합하여 lq 지수 테이블 산출
food_service_lq_df = food_service_df[['dong', 'quarter', 'food_service_lq']].drop_duplicates()

retail_lq_df = retail_df[['dong', 'quarter', 'retail_lq']].drop_duplicates()

franchise_lq_df = df_gentrification[['dong', 'quarter', 'franchise_lq']].drop_duplicates()

youth_sales_lq_df = df_gentrification[['dong', 'quarter', 'youth_sales_lq']].drop_duplicates()

net_growth_rate_df = df_gentrification[['dong', 'quarter', 'total_net_growth_rate']].drop_duplicates()

final_lq_df = food_service_lq_df.merge(
    retail_lq_df, on=['dong', 'quarter'], how='outer'
).merge(
    franchise_lq_df, on=['dong', 'quarter'], how='outer'
).merge(
    youth_sales_lq_df, on=['dong', 'quarter'], how='outer'
).merge(
    net_growth_rate_df, on=['dong', 'quarter'], how='outer'
)
```
