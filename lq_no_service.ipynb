```python
import pandas as pd

# Load the uploaded file
file_path = 'table/최종 데이터(no_service).csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
data['rent_lq'] = (data['monthly_rent_per_pyeong']) / (data['monthly_rent_seoul'])

data['rent_lq']
```

```python
data['seoul_total_fac_cnt'] = data.groupby(['quarter'])['visitor_facility_count'].transform('sum')
data['dong_total_fac_cnt'] = data.groupby(['quarter', 'dong'])['visitor_facility_count'].transform('sum')
data['pop_total_seoul'] = data.groupby('quarter')['pop_total'].transform('sum')

data['facility_lq'] = (data['dong_total_fac_cnt'] / data['pop_total']) / (
    data['seoul_total_fac_cnt'] / data['pop_total_seoul'])
```

```python
data.head()
```

```python
data=data.drop(columns=['pop_male', 
                   'pop_female', 
                   'visitor_facility_count',
                   'avg_monthly_income',
                   'total_expenditure', 
                   'monthly_rent_per_pyeong',
                   'monthly_rent_seoul',
                   'seoul_total_fac_cnt',
                   'dong_total_fac_cnt',
                   'pop_total_seoul'])
```

```python
data.to_csv('table/lq_no_service.csv', index=False, encoding='utf-8')
```
